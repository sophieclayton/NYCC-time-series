```python
%matplotlib inline
import numpy as np
import pandas as pd
import glob
```

```python
# get the list of hdr files

hdr_files = glob.glob('../data/IFCB_DATA/*.hdr')

f = open(hdr_files[0], "r")
hdr = f.read()
f.close()

lookup = 'runTime:'
lookup2 = 'inhibitTime:'

orig_id = []
rt = []  
it = []

for h in hdr_files:
    with open (h, 'rt') as myfile: 
        orig_id.append('_MOORING_'+ h[18:-4])
        for line in myfile: 
            if lookup in line:# For each line, stored as myline,
                rt.append(np.float(line[9:20]))  
            if lookup2 in line:
                it.append(np.float(line[13:20]))

ifcb_vols = pd.DataFrame()
ifcb_vols['orig_id'] = orig_id
ifcb_vols['runTime'] = rt
ifcb_vols['inhibitTime'] = it
ifcb_vols['vol_ml'] = 0.25*(np.array(rt) - np.array(it))/60
```
